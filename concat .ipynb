```python
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
dataset = pd.read_csv('User_Data.csv')
x = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
type(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
x_train2 = mn.fit_transform(x_train)
x_test2 = mn.transform(x_test)
```

```python
sc = StandardScaler()
x_train1 = sc.fit_transform(x_train)
x_test1 = sc.transform(x_test)
```

```python
print(x_train1)
```

```python
print(x_train2)
```

```python
dff=pd.concat([pd.DataFrame(x_train1,columns=['a','b']),pd.DataFrame(x_train2,columns=['a','b'])],axis=1)
dff
```

```python
print(x_test1)
```

```python
dff1=pd.concat([pd.DataFrame(x_test1,columns=['a','b']),pd.DataFrame(x_test2,columns=['a','b'])],axis=1)
dff1
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(dff, y_train)
```

```python
y_pred = classifier.predict(dff1)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix : \n",cm)
```
